```python
import matplotlib.pyplot as plt
import numpy as np
from math import pi


r0 = 15
rf = 65/2
D = 1
t = np.linspace(0,25,1000)
r = r0 + (D/2)*((t/pi) + 1)
x = r*np.cos(t)
y = r*np.sin(t)
plt.plot(x,y)
plt.axis('equal')

```

```python
tf = pi*((rf-r0)*2/D -1)
print(tf)
```

```python
l = r0*tf + (D/2)*(tf + (tf**2)/(2*pi))
print(l)
```
